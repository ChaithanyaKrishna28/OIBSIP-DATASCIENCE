## Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
car = ['fueltype','aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation','enginetype','cylindernumber','fuelsystem']
for var in car:
    label_encoder = LabelEncoder()
    df[var+'_label_encoded'] = label_encoder.fit_transform(df[var])
df.drop(car,axis=1,inplace = True)
print("The encoded data\n:")
df
```

# splitting the data (80-20)

```python
#we must drop the price because we have to predict it
x = df.drop(columns =['price','CarName'],axis=1)
y = df['price']
```

```python
print("The x data:\n")
x
```

```python
print('The prices data:\n')
y
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

## TRAINING DATA 

```python
print("TRAINING DATA")
print("X_train")
display(X_train)
print("Y_train")
display(Y_train)
```

## TESTING DATA

```python
print("TESTING DATA")
print("X_test")
display(X_test)
print("Y_test")
display(Y_test)
```

## Model training


```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

## Model Evaluation

```python
#training set prediction
training_data_prediction = model1.predict(X_train)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared error :",error_score)
```

## visualize the actual prices and predicted price 

```python
plt.scatter(Y_train,training_data_prediction,color = 'red')
plt.xlabel('Actual Price')
plt.ylabel('predictied Price')
plt.title("Actual price V/s Predicted price")
plt.show()
```

```python
#for the test data predidctions
test_data_prediction = model1.predict(X_test)
```
